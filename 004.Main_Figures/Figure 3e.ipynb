```r
library(pheatmap)
library(RColorBrewer)
library(glue)
```

```r
data_dir = '/data/yuchen_data/Stereo_seq/All_Figures/Figure3_v2/'
```

```r
df = read.table(glue('{data_dir}/Cytokines_and_Effector_molecules_with_mature.txt'), row.names=1, header=T, sep='\t')
df = df[, c('Pre_Conforming_Tcell', 'Post_Conforming_Tcell', 'Pre_Deviating_Tcell', 'Post_Deviating_Tcell')]
df = t(df)
df = scale(df)
df = data.frame(t(df))
df[df > 1.5] = 1.5
df[df < -1.5] = -1.5
df =  df[rev(order(df$Post_Conforming_Tcell)), ]
pheatmap(df, cluster_cols=F, cluster_row=F, scale='none', gaps_col=c(2))
```
